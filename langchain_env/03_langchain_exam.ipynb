```python
from langchain_openai import OpenAIEmbeddings

embeddings_model = OpenAIEmbeddings(
    model="text-embedding-3-small"
)
embeddings_model
```

```python
embeddings_model.embedding_ctx_length
```

```python
documents = [
    "ì¸ê³µì§€ëŠ¥ì€ ì»´í“¨í„° ê³¼í•™ì˜ í•œ ë¶„ì•¼ì…ë‹ˆë‹¤.",
    "ë¨¸ì‹ ëŸ¬ë‹ì€ ì¸ê³µì§€ëŠ¥ì˜ í•œ ë¶„ì•¼ì…ë‹ˆë‹¤.",
    "ë”¥ëŸ¬ë‹ì€ ë¨¸ì‹ ëŸ¬ë‹ì˜ í•œ ì¢…ë¥˜ì…ë‹ˆë‹¤.",
    "ìì—°ì–´ ì²˜ë¦¬ëŠ” ì»´í“¨í„°ê°€ ì¸ê°„ì˜ ì–¸ì–´ë¥¼ ì´í•´í•˜ê³  ìƒì„±í•˜ëŠ” ê¸°ìˆ ì…ë‹ˆë‹¤.",
    "ì»´í“¨í„° ë¹„ì „ì€ ì»´í“¨í„°ê°€ ë””ì§€í„¸ ì´ë¯¸ì§€ë‚˜ ë¹„ë””ì˜¤ë¥¼ ì´í•´í•˜ëŠ” ë°©ë²•ì„ ì—°êµ¬í•©ë‹ˆë‹¤.",
]

document_embeddings = embeddings_model.embed_documents(documents)

print(f'ì„ë² ë”© ë²¡í„°ì˜ ê°œìˆ˜: {len(document_embeddings)}')
print(f'ì„ë² í•œ ë²¡í„°ì˜ ì°¨ì›: {len(document_embeddings[0])}')

print(document_embeddings[0])
```

```python
embedded_query = embeddings_model.embed_query("ì¸ê³µì§€ëŠ¥ì´ë€ ë¬´ì—‡ì¸ê°€ìš”?")


# print(f'ì„ë² ë”© ë²¡í„°ì˜ ê°œìˆ˜: {len(document_embeddings)}')
print(f'ì„ë² í•œ ë²¡í„°ì˜ ì°¨ì›: {len(embedded_query)}')
print(embedded_query)
```

```python
from langchain_community.utils.math import  cosine_similarity
import numpy as np

def find_most_similar(query, document_embeddings):
    query_embedding = embeddings_model.embed_query(query)
    similarities = cosine_similarity(
        [query_embedding], document_embeddings
    )[0]
    most_similar_idx = np.argmax(similarities) #ê·¸ ì¤‘ ì œì¼í° ê²ƒì˜ ì¸ë±ìŠ¤ë¥¼ ì•Œë ¤ì¤Œ
    return documents[most_similar_idx], similarities[most_similar_idx]
        #5ê°œ ë¬¸ì„œì¤‘ ìœ ì‚¬ë„ ê°€ì¥ ë†’ì€ ë¬¸ì„œ, ìœ ì‚¬ë„ ì ìˆ˜ 
        
queries = [
    " ì¸ê³µì§€ëŠ¥ì´ë€ ë¬´ì—‡ì¸ê°€ìš”?",
    " ë”¥ëŸ¬ë‹ê³¼ ë¨¸ì‹ ëŸ¬ë‹ì˜ ê´€ê³„ëŠ” ì–´ë–»ê²Œ ë˜ë‚˜ìš”?",
    " ì»´í“¨í„°ê°€ ì´ë¯¸ì§€ë¥¼ ì´í•´í•˜ëŠ” ë°©ë²•ì€?",
]

for query in queries:
    most_similar_doc, similarity = find_most_similar(query, document_embeddings)
    print(f'ì§ˆë¬¸: {query}')
    print(f'ê°€ì¥ ìœ ì‚¬í•œ ë¬¸ì„œ: {most_similar_doc}')
    print(f'ìœ ì‚¬ë„: {similarity:.4f}')
    print('-'*50)
```
